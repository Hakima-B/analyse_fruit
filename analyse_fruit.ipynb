```python

print("\nüì¶ √âTAPE 1 : Chargement des outils...")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("‚úÖ Tous les outils sont pr√™ts !")
```

```python
print("\nüçì √âTAPE 2 : Cr√©ation de nos donn√©es de fruits...")

# Donn√©es d'un petit magasin de fruits
fruits_data = {
    'Fruit': ['Pomme', 'Banane', 'Orange', 'Fraise', 'Kiwi', 'Mangue', 'Ananas', 'Poire', 'Peche'],
    'Prix_kg': [4.00, 2.80, 4.20, 8.50, 5.60, 6.30, 4.80, 3.90, 3.95],
    'Ventes_kg': [45, 67, 32, 23, 18, 15, 12, 28, 30],
    'Couleur': ['Rouge', 'Jaune', 'Orange', 'Rouge', 'Vert', 'Jaune', 'Jaune', 'Vert', 'Orange'],
    'Saison': ['Automne', 'Toute_annee', 'Hiver', 'Printemps', 'Automne', 'Ete', 'Ete', 'Automne', 'Printemps']
}
```

```python
# Cr√©er notre tableau pandas
df = pd.DataFrame(fruits_data)

print(f"üìä Nous avons {len(df)} types de fruits dans notre magasin")
print("\nüîç Voici nos donn√©es :")
print(df)
```

```python
print("\n\nüîç √âTAPE 3 : PREMI√àRE ANALYSE")
print("-" * 40)

# Question 1 : Quel est le fruit le plus cher ?
fruit_plus_cher = df.loc[df['Prix_kg'].idxmax()]
print(f"üèÜ Fruit le plus cher : {fruit_plus_cher['Fruit']} √† {fruit_plus_cher['Prix_kg']}‚Ç¨/kg")

# Question 2 : Quel fruit se vend le mieux ?
fruit_plus_vendu = df.loc[df['Ventes_kg'].idxmax()]
print(f"üìà Fruit le plus vendu : {fruit_plus_vendu['Fruit']} avec {fruit_plus_vendu['Ventes_kg']} kg")

# Question 3 : Calculer le chiffre d'affaires de chaque fruit
df['Chiffre_affaires'] = df['Prix_kg'] * df['Ventes_kg']
print(f"\nüí∞ Chiffre d'affaires total du magasin : {df['Chiffre_affaires'].sum():.2f}‚Ç¨")

# Question 4 : Prix moyen des fruits
prix_moyen = df['Prix_kg'].mean()
print(f"üí∏ Prix moyen des fruits : {prix_moyen:.2f}‚Ç¨/kg")

# Afficher le tableau avec les chiffres d'affaires
print("\nüìã Tableau complet avec chiffres d'affaires :")
print(df[['Fruit', 'Prix_kg', 'Ventes_kg', 'Chiffre_affaires']].round(2))
```

```python
print("\n\nüìä √âTAPE 4 : VISUALISATIONS")
print("-" * 40)

# Cr√©ation de 4 graphiques
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('üçé ANALYSE DE NOTRE MAGASIN DE FRUITS', fontsize=16, fontweight='bold')

# Graphique 1 : Prix des fruits
axes[0, 0].bar(df['Fruit'], df['Prix_kg'], color='lightcoral')
axes[0, 0].set_title('üí∞ Prix des fruits (‚Ç¨/kg)')
axes[0, 0].set_ylabel('Prix (‚Ç¨)')
axes[0, 0].tick_params(axis='x', rotation=45)

# Graphique 2 : Ventes des fruits
axes[0, 1].bar(df['Fruit'], df['Ventes_kg'], color='lightblue')
axes[0, 1].set_title('üì¶ Quantit√©s vendues (kg)')
axes[0, 1].set_ylabel('Ventes (kg)')
axes[0, 1].tick_params(axis='x', rotation=45)

# Graphique 3 : Chiffre d'affaires par fruit
axes[1, 0].bar(df['Fruit'], df['Chiffre_affaires'], color='lightgreen')
axes[1, 0].set_title('üí∏ Chiffre d\'affaires par fruit (‚Ç¨)')
axes[1, 0].set_ylabel('Chiffre d\'affaires (‚Ç¨)')
axes[1, 0].tick_params(axis='x', rotation=45)

# Graphique 4 : Relation Prix vs Ventes
axes[1, 1].scatter(df['Prix_kg'], df['Ventes_kg'], s=100, color='orange', alpha=0.7)
axes[1, 1].set_title('üìà Prix vs Ventes')
axes[1, 1].set_xlabel('Prix (‚Ç¨/kg)')
axes[1, 1].set_ylabel('Ventes (kg)')

# Ajouter les noms des fruits sur le scatter plot
for i, fruit in enumerate(df['Fruit']):
    axes[1, 1].annotate(fruit, (df['Prix_kg'][i], df['Ventes_kg'][i]), 
                       xytext=(5, 5), textcoords='offset points', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
print("\n\nüåà √âTAPE 5 : ANALYSE PAR COULEUR")
print("-" * 40)

# Grouper par couleur
analyse_couleur = df.groupby('Couleur').agg({
    'Ventes_kg': 'sum',
    'Chiffre_affaires': 'sum',
    'Prix_kg': 'mean'
}).round(2)

print("üìä Ventes par couleur de fruit :")
print(analyse_couleur)

# Graphique des ventes par couleur
plt.figure(figsize=(8, 6))
couleurs = ['green', 'orange', 'red', 'gold']  # Couleurs pour le graphique
plt.pie(analyse_couleur['Ventes_kg'], 
        labels=analyse_couleur.index, 
        autopct='%1.1f%%',
        colors=couleurs,
        startangle=90)
plt.title('üåà R√©partition des ventes par couleur de fruit', fontsize=14, fontweight='bold')
plt.axis('equal')
plt.show()
```

```python
print("\n\nüí° √âTAPE 6 : CONSEILS POUR AM√âLIORER LE MAGASIN")
print("-" * 50)

# Fruit le plus rentable
fruit_plus_rentable = df.loc[df['Chiffre_affaires'].idxmax()]
print(f"ü•á Fruit le plus rentable : {fruit_plus_rentable['Fruit']} ({fruit_plus_rentable['Chiffre_affaires']:.2f}‚Ç¨)")

# Fruits peu vendus mais chers
fruits_chers_peu_vendus = df[(df['Prix_kg'] > prix_moyen) & (df['Ventes_kg'] < 20)]
if not fruits_chers_peu_vendus.empty:
    print(f"\n‚ö†Ô∏è  Fruits chers mais peu vendus (√† promouvoir) :")
    for _, fruit in fruits_chers_peu_vendus.iterrows():
        print(f"   ‚Ä¢ {fruit['Fruit']} : {fruit['Prix_kg']}‚Ç¨/kg, seulement {fruit['Ventes_kg']}kg vendus")

# Fruits populaires √† prix abordable
fruits_populaires = df[(df['Ventes_kg'] > 30) & (df['Prix_kg'] < prix_moyen)]
if not fruits_populaires.empty:
    print(f"\nüéØ Fruits populaires et abordables (√† maintenir en stock) :")
    for _, fruit in fruits_populaires.iterrows():
        print(f"   ‚Ä¢ {fruit['Fruit']} : {fruit['Prix_kg']}‚Ç¨/kg, {fruit['Ventes_kg']}kg vendus")

print(f"\nüí∞ R√âSUM√â FINANCIER :")
print(f"   ‚Ä¢ Chiffre d'affaires total : {df['Chiffre_affaires'].sum():.2f}‚Ç¨")
print(f"   ‚Ä¢ Ventes totales : {df['Ventes_kg'].sum()} kg")
print(f"   ‚Ä¢ Panier moyen : {(df['Chiffre_affaires'].sum() / df['Ventes_kg'].sum()):.2f}‚Ç¨/kg")
```
